```python
import random
from IPython.display import clear_output
```

```python
asset = 10000
login_status = False
```

```python
def login():
    global login_status
    print('---GAME LÔ ĐỀ HỌC---')
    while True:
        username = input('Nhập username: ')
        password = input('Nhập password: ')
        if(username == 'admin' and password == 'admin'):
            login_status = True
            break
        print('Sai tên người dùng hoặc mật khẩu, vui lòng thử lại.')
```

```python
def roll():
    first = '{:05d}'.format(random.randint(0, 99999))
    second = '{:05d}'.format(random.randint(0, 99999))
    third = '{:05d}'.format(random.randint(0, 99999))
    fourth = '{:05d}'.format(random.randint(0, 99999))
    fifth = '{:05d}'.format(random.randint(0, 99999))
    sixth = '{:05d}'.format(random.randint(0, 99999))
    seventh = '{:05d}'.format(random.randint(0, 99999))
    list_prize = {
        'Giải nhất': first,
        'Giải nhì ': second,
        'Giải ba  ': third,
        'Giải tư  ': fourth,
        'Giải năm ': fifth,
        'Giải sáu ': sixth,
        'Giải bảy ': seventh,
    }
    return list_prize
```

```python
def print_prize(list_prize):
    print('-'*20)
    print('---Danh sách giải---')
    for key, value in list_prize.items():
        print(f'{key}: {value}')
    print('-'*20)

print_prize(roll())
```

```python
def bet_input():
    global asset
    while True:
        bet = input("Nhập tiền cược mà bạn muốn cược (tiền cược < tổng tiền bạn có): ")
        if bet.isdigit() and 0<int(bet)<asset:
            return int(bet)
        print("Nhập sai định dạng hoặc quá tổng tiền. Nhập lại!")

def bet_num_input():
    while True:
        bet_nums = input("Nhập số lô mà bạn muốn cược (10-99): ").split(',')
        if all(bet_num.isdigit() and 10<=int(bet_num)<=99 for bet_num in bet_nums):
            return bet_nums
        print("Nhập sai định dạng, vui lòng nhập lại!")
```

```python
def process(bet, bet_nums, prize_list):
    global asset
    won_count = 0
    won_nums = []

    for prize, value in prize_list.items():
        for bet_num in bet_nums:
            if bet_num == value[-2:]:
                won_count += 1
                if bet_num not in won_nums:
                    won_nums.append(bet_num)

    if won_count:
        earn = bet * won_count * 70
        lost = bet * (len(bet_nums) - len(won_nums))
        asset += earn - lost
        print(f'Bạn đã trúng {won_count} nháy!')
        print('Số lô bạn trúng là:', ', '.join(list(map(str, won_nums))))
        print(f'Số tiền bạn trúng là {earn}')
        print(f'Số tiền bạn thua là {lost}')
        print(f'Tổng tiền bạn có là {asset}')
    else:
        lost = bet * len(bet_nums)
        asset -= lost
        print('Bạn đã thua lô!')
        print(f'Số tiền bạn thua là {lost}')
        print(f'Tổng tiền bạn còn là {asset}')
```

```python
def play():
    while True:
        bet = bet_input()
        bet_nums = bet_num_input()
        print(f'Các số lô bạn chơi là {bet_nums}')

        prize_list = roll()
        print_prize(prize_list)

        process(bet, bet_nums, prize_list)
        while True:
            ctn = input("Bạn có muốn chơi tiếp không (0: không, 1: có): ")
            if(ctn=='1' or ctn=='0'): break
            print("Nhập không hợp lệ!")
        if(ctn=='0'):
            break
```

```python
def pay():
    global asset
    while True:
        while True:
            pay_amount = input("Nhập số tiền mà bạn muốn nạp: ")
            if pay_amount.isdigit(): break
            print("Nhập không hợp lệ!")
        asset += int(pay_amount)
        print(f"Tổng tiền sau khi bạn nạp là {asset}")
        while True:
            ctn = input("Bạn có muốn nạp tiền tiếp không (0: không, 1: có): ")
            if(ctn=='1' or ctn=='0'): break
            print("Nhập không hợp lệ!")
        if(ctn=='0'):
            break
```

```python
def menu():
    print("GAME LÔ ĐỀ HỌC")
    print("1. Chơi lô")
    print("2. Nạp tiền")
    print("3. Đăng xuất")
```

```python
def main():
    global asset
    global login_status
    if not login_status:
        login()
        clear_output()
    menu()
    while True:
        choice = input('Lựa chọn của bạn: ')
        if(choice=='1' or choice=='2' or choice=='3'): break
        print('Nhập sai lựa chọn, vui lòng nhập lại !')
    if choice=='1':
        clear_output()
        play()
        clear_output(wait=True)
        main()
    elif choice=='2':
        clear_output()
        pay()
        clear_output(wait=True)
        main()
    else:
        login_status = False
        print("Chào bạn, hẹn gặp lại")
```

```python
main()
```
